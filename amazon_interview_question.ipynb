```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

import pandas as pd

x = np.linspace(0,2*30**0.5,201)

y = 20+(30**0.5-x)**2+(20**0.5-x)**2

fig = go.Figure(go.Scatter(
    mode = "lines+markers",
    x = x,
    y = y))

fig.update_xaxes(title_text = 'X values')
fig.update_yaxes(title_text = 'Y values')

fig.show()
```

```python
ll_ = []

#a second order equation of x, 
lambdf = lambda x,A,B : A*x**2 + B*x + 50

for A in np.linspace(-10,10,301):
    for B in np.linspace(-10,10,201):
        for x in np.linspace(0,100,601):

            xp_ = x/2                 #middle point
            ff1 = B+2*A*xp_           #derivative of the labmdf @xp, must be equal to "zero"
            ff2 = A*xp_**2+B*xp_+50   #calculates height of the xp, must be equal to 20 
            yright = A*x**2+B*x+50

            lrope = 0                 #holds rope lenght in each iteration

            if ff1<1 and ff1>-1 and ff2<25 and ff2>15 and yright<55 and yright>50:

                xpp = np.linspace(0,x,250)
                
                ypp = lambdf(xpp, A, B)

                xpf = xpp[1:]-xpp[0:-1]
                ypf = ypp[1:]-ypp[0:-1]

                lrope = np.sum((xpf**2+ypf**2)**0.5)

                if lrope<84 and lrope>76:

                    ll_.append([round(A,3),round(B,3),round(x,3),round(ff1,3),round(ff2,3), np.round(lrope,3), np.round(yright,3)])
    
```

```python
ll_ = []

#a second order equation of x, 
lambdf = lambda x,A,B : A*x**2 + B*x + 50

for A in np.linspace(Amin,Amax,401):
    for B in np.linspace(Bmin,Bmax,401):
        for x in np.linspace(xmin,xmax,801):

            xp_ = x/2                 #middle point
            ff1 = B+2*A*xp_           #derivative of the labmdf @xp, must be equal to "zero"
            ff2 = A*xp_**2+B*xp_+50   #calculates height of the xp, must be equal to 20 
            yright = A*x**2+B*x+50

            lrope = 0                 #holds rope lenght in each iteration

            if ff1<0.5 and ff1>-0.5 and ff2<20.5 and ff2>19.5 and yright<50.2 and yright>49.8:

                xpp = np.linspace(0,x,250)
                
                ypp = lambdf(xpp, A, B)

                xpf = xpp[1:]-xpp[0:-1]
                ypf = ypp[1:]-ypp[0:-1]

                lrope = np.sum((xpf**2+ypf**2)**0.5)

                if lrope<80.5 and lrope>79.5:

                    ll_.append([round(A,3),round(B,3),round(x,3),round(ff1,3),round(ff2,3), np.round(lrope,3), np.round(yright,3)])
```
